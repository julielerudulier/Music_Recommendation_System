```python
# Import packages
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import data
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
# Take a look at the data
df.head()
```

```python
# Set 'Experience Years' as our predictor and 'Salary' as our response variable
X = df[['Experience Years']]
y = df['Salary']
```

```python
# Instantiate the model
model = LinearRegression()
```

```python
# Fit the model
model.fit(X, y)
```

```python
# Get the model's coefficient, Beta1
model.coef_
```

```python
# Get the model's intercept, Beta0
model.intercept_
```

```python
# Predict values for any given x
y_pred = model.predict(X)
```

```python
# Plot the original data and the regression line, i.e. our model
ax = sns.scatterplot(data=df[['Experience Years', 'Salary']], x='Experience Years', y='Salary', alpha=0.8)
ax = plt.plot(X, y_pred, color='red')
plt.legend(labels=["Data", "Model"])
plt.show()
```

```python
# Compute the R-squared score
r2 = r2_score(y, y_pred)
print(round(r2, 3))
```
